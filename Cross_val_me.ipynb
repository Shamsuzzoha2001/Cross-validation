```python
pos=df[df["isFraud"]==1].copy()
neg=df[df["isFraud"]==0].sample(100000)
dat=pd.concat([pos,neg])
```

```python
dat.drop(columns=df.columns[[0,1]],inplace=True)
```

```python
dat
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score
from sklearn.ensemble import RandomForestClassifier
```

```python
features=dat.columns[[2,4,5,7,8]]
features
```

```python
X=dat[features]
y=dat["isFraud"]

# to train the model
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.7,random_state=90)


#to train the model
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_predicts=model.predict(x_test)
average_precision_score(y_predicts,y_test)
```

# Cross-validation

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,X,y,cv= 5)
print("cross validation scores :", scores)
print("mean of scores:",np.mean(scores))
```
